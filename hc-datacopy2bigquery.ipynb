```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
inputdatapath    = '/kaggle/input/home-credit-credit-risk-model-stability/parquet_files/'
gc_project_id    = 'kagglehomecredit'
bq_dataset_name  = 'homecredit_raw_test'
```

```python
from google.cloud import bigquery
bq_client = bigquery.Client(project=gc_project_id)
```

```python
from kaggle_datasets import KaggleDatasets
kaggle_dataset_path = KaggleDatasets().get_gcs_path() 
print(kaggle_dataset_path)
```

```python
bq_client.create_dataset(bq_dataset_name, exists_ok=True)
```

```python
def load_into_bq(uri, gc_project_id, bq_client, bq_dataset_name, table_name):
    
    table_ref  = bq_client.dataset(bq_dataset_name).table(table_name)
    job_config = bigquery.LoadJobConfig(
#                     autodetect = True,
#                     skip_leading_rows = 1,
                    source_format=bigquery.SourceFormat.PARQUET,
            )
    
    load_job = bq_client.load_table_from_uri(uri, table_ref, job_config=job_config)
    print(f"Loading data from {uri} to BigQuery table {table_name}...")

    # Wait for the job to complete
    load_job.result()
```
